b. Traza un histograma y una ovija. Analízalas.


```python
marcas_clase = [1, 2, 3, 4, 5, 6] # Datos del eje x
frecuencias = [9.375, 6.25, 25, 21.875, 28.125, 9.375] # Datos del eje y
frecuencias = [9.375, 15.625, 40.625, 62.5, 90.625, 100] # Datos del eje y

# Ajustes necesarios para trazar el poligono (NO TOCA)
datos_x=[0]+marcas_clase
datos_y=[0]+frecuencias
marcas_texto = ["9.375", "15.625", "40.625", "62.5", "90.625", "100"]
plt.figure(figsize = (12,6)) # Ancho, alto de la figura
plt.plot(datos_x, datos_y, 
         linewidth = 5, color = "r", linestyle=":", 
         marker = "v", markersize = 15, markerfacecolor = "y", markeredgecolor = "b")
        
# Esta ojiva cuenta con: 
# - Linea punteada (linestyle=":") https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# - Marcador triangular (marker = "v") https://matplotlib.org/stable/api/markers_api.html
# - 

        
plt.xticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 45) # Colocar textos sobre el eje x
plt.xlabel("Marcas de clase", fontsize = 20) # Etiqueta del eje x
plt.ylabel("Frecuencias", fontsize = 20) # Etiqueta del eje y
plt.title("Frecuencia acumulada", fontsize = 30) # Titulo de grafico
plt.grid() # Se activa la cuadricula sobre el 
plt.show() # Se muestra el grafico
```

```python
marcas_clase = [1, 2, 3, 4, 5, 6]

frecuencias = [9.375, 6.25, 25, 21.875, 28.125, 9.375]

plt.figure(figsize = (12,6))

marcas_texto = ["9.375", "6.25", "25", "21.875", "28.125", "9.375"]

plt.bar(marcas_clase, frecuencias, 
        
        
        
       width = 1, edgecolor = "k", 
        
       color = ["#EAE279", "#D1FE96", "#96FEC4", "#96E8FE", "#96A9FE", "#FE96E8"])

plt.xticks(marcas_clase, marcas_texto, fontsize = 15, rotation = 45)

plt.xlabel("Marcas de clase", fontsize = 20)

plt.ylabel("Frecuencias", fontsize = 20)

plt.title("Frecuencias absolutas", fontsize = 30)

plt.show() 
```
